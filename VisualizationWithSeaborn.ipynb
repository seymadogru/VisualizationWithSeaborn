Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
df.head()
```

```python
covid= df[df.location.isin(['Spain','France','Germany','Italy'])]
my_covid = covid[covid['date']>'2020-04-01']

plt.figure(figsize=(5, 3), dpi = 100)

sns.lineplot(x = 'date', y = 'total_deaths', data = my_covid, hue = 'location')
plt.xticks(rotation = 90)
plt.show()
```

Plot a bar plot with seaborn for average death number that compares continents.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
a = pd.DataFrame(df.groupby(by='continent')['total_deaths'].mean())

b = a.reset_index()

plt.figure(figsize=(9, 6), dpi = 100)
sns.barplot(x = 'continent', y = 'total_deaths', data = b)
plt.ylabel('total deaths avarage')
plt.xticks(rotation = 90)
plt.show()
```

Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and kde arguments.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
plt.figure(figsize=(12, 4), dpi = 100)

covid= df[df.location == 'Italy']

plt.subplot(1,2,1)
sns.distplot(covid.new_cases.dropna(),kde = False, color = '#FF914D',bins=20)
plt.title("kde = False , bins =20",color='r')
plt.xticks(rotation = 90)

plt.subplot(1,2,2)
sns.distplot(covid.new_cases.dropna(), kde = False, color = '#22B573',bins=50)
plt.title("kde = False , bins = 50",color='r')
plt.xticks(rotation = 90)

plt.figure(figsize=(12, 4), dpi = 100)

plt.subplot(1,2,1)
sns.distplot(covid.new_cases.dropna(), kde = True, color = '#FF914D',bins=50)
plt.title("kde = True,bins=50",color='r')
plt.xticks(rotation = 90)

plt.subplot(1,2,2)
sns.distplot(covid.new_cases.dropna(), kde = True, color = '#22B573',bins=20)
plt.title("kde = True,bins=20",color='r')
plt.xticks(rotation = 90)

plt.tight_layout()
plt.show()
```

Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
plt.figure(figsize=(12, 4), dpi = 100)
covid= df[df.location.isin(['Spain','France'])]
plt.subplot(131)
sns.boxplot(x = 'location',  y = 'new_deaths', data = covid, palette = ['#FF914D', '#22B573'])
plt.title("boxplot")

plt.subplot(132)
sns.swarmplot(x = 'location',  y = 'new_deaths', data = covid, palette = ['#FF914D', '#22B573'])
plt.title("swarmplot")

plt.subplot(133)
sns.violinplot(x = 'location',  y = 'new_deaths', data = covid, palette = ['#FF914D', '#22B573'])
plt.title("violinplot")

plt.tight_layout()
plt.show()
```
